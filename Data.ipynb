```python
import pandas as pd
```

```python
sales_0= pd.read_csv("data\daily_sales_data_0.csv")
sales_1= pd.read_csv("data\daily_sales_data_1.csv")
sales_2= pd.read_csv("data\daily_sales_data_2.csv")
```

```python
sales_1.head()
```

```python
sales_2.head()
```

```python
sales=pd.concat([sales_0,sales_1, sales_2], ignore_index=True)
sales.info(), sales["product"].value_counts()
```

```python
sales = sales[sales["product"]== "pink morsel"]
```

```python
sales.info(), sales["product"].value_counts()
```

```python
sales['price'] = sales['price'].str.replace('$', '', regex=False)
sales['price'] = pd.to_numeric(sales['price'], errors='coerce')
```

```python
sales['sales'] = sales['price'] * sales['quantity']
```

```python
daily_sales_data = sales.copy()
daily_sales_data.to_csv(r"data\daily_sales_data.csv")
```
